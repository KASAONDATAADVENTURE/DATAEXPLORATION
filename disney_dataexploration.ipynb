```python
import pandas as pd

#Loading the dataset

disney_data = pd.read_csv('Disney_data_cleaned.csv')
```

```python
import pandas as pd

# Display basic information about the dataset
print("\nBasic information about the dataset:")
print(disney_data.info())
```

```python
import pandas as pd

# Summary statistics of numerical variables
print("\nSummary statistics of numerical variables:")
print(disney_data.describe())
```

```python
import pandas as pd

# Frequency table for rating variable to understand how many shows have been released with what rating
print("\nFrequency table for the 'rating' variable:")
print(disney_data['rating'].value_counts())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Grouping data by release year and counting titles in each year
titles_by_year = disney_data.groupby('release_year')['title'].count()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
titles_by_year.plot(kind='bar')
plt.xlabel('Release Year')
plt.ylabel('Count of Titles')
plt.title('Distribution of Titles by Release Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
